```python
df = pd.read_csv('http://bit.ly/uforeports')
```

```python
df.head()
```

目標：統計每州UFO出現次數

```python
# 依項目集合值：dataframe名稱.groupby('項目')
# 計算次數：dataframe名稱.count()
df_state = df.groupby('State').count()
df_state.head()
```

Pandas傾向不變更原始資料，指令會產生新的datafrme

```python
df_state.sort_values(by='Time', ascending=False)
```

```python
df_state.head(10)
```

```python
df_state.sort_values(by='Time', ascending=False, inplace=True)
```

```python
df_state.head()
```

```python
df_state[:10].Time.plot() #取前十筆畫圖
```

```python
df_state[:10].Time.plot(kind='bar') #取前十筆畫圖(bar:長條圖)
```

```python
import pandas_datareader.data as web
```

有財務、股票相關的實際資料

```python
# web.DataReader(股票代號, 資料源, 開始時間, 結束時間) 
df = web.DataReader("AAPL", "yahoo", start='2012-9-1', end='2017-8-31') #讀入蘋果股票
df.head()
```

```python
P = df['Adj Close'] #調整收盤價
P.head() #只有一欄資料，所以是Series
```

```python
P.plot()
```

```python
# 計算報酬率：(今日報酬率-昨日報酬率)/昨日報酬率
P.diff() #與前一期差別
```

```python
r = P.diff()/P #報酬率
```

```python
r.plot()
```

```python
r[-100:].plot() #最新100天
```

```python
#移動平均：讓漲幅變平滑，好看出走向
P.rolling(window=20).mean() #window：多少範圍的平均，每20期開個小視窗
```

```python
P.plot()
P.rolling(window=20).mean().plot()
P.rolling(window=60).mean().plot()
```
